```python
#The triangle function...
t = Turtle()     # creates our turtle, named t
t.speed(10)      # from 1 to 10


def tri(n):
    """Draws n 100-pixel sides of an equilateral triangle.
       Note that n doesn't have to be 3 (!)
    """
    if n == 0:
        return      # No sides to draw, so stop drawing
    else:
        t.forward(100)
        t.left(120)
        tri(n-1)    # Recur to draw the rest of the sides!

        
tri(3)   # this runs the function with an input of 3
```

```python
#
# a generic polygon example
#

# here, we DO define a new turtle, 
# so this will run below this cell in its own space 
# [later examples use new turtles, too...]

t = Turtle()
t.speed(10)
t.setbgColor('white')
t.pensize(3)
t.pencolor("DarkGreen")

t.penup()    # get setup in the lower left w/o drawing...
t.backward(100)
t.left(90)
t.backward(100)
t.right(90)
t.pendown()  # place pen back on the canvas


def poly(n,N):
    """  n == sides to go (to be drawn) [[varies]]
         N == total # of sides in the regular polygon [[constant]]
    """
    if n == 0:  
        return  # stop!
    else:
        # print("side", n)
        t.forward(100)
        angle = 360/N
        t.left(angle)
        poly(n-1,N)   # draw the remaining sides...
        
poly(9,9)
```

```python
#The chai function...
t = Turtle()
t.speed(10)
t.pensize(3)

def chai(size):
    """Our chai function!"""
    if (size < 5): 
        return
    else:
        t.forward(size)
        t.left(90)
        t.forward(size/2)
        t.right(90)
        # room to recur
        t.right(90)
        t.forward(size)
        t.left(90)
        # room to recur
        t.left(90)
        t.forward(size/2)
        t.right(90)
        t.backward(size)
        return
    
# Let's call chai(100)
chai(100)
```
